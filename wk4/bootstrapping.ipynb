## Bootstrap sampling
Bootstrp sampling is a method that involves drawing of sample data repeatedly with replacement from a data source to estimate a population parameter.

For example, if our population size is 1000, then to find the mean, instead of considering all 1000 entries we can take 50 samples each of 4 entries and calculate the mean of each sample. This way we will be taking the average of 200 entries (50 X 4) chosen randomly. This technique is particularly useful for research in huge populations. 

```python
# Imports

import numpy as np
import random
```

```python
# generate random data

x = np.random.normal(loc=300.0, size=1000)
```

#####  Calculate the mean of this data set
This is the actual mean of the population

```python
print(np.mean(x))
```

##### Create 50 samples of size 4 each to estimate the mean. If you run this code again and again you'll see that you get a different value for `sample_mean` each time you run it.

```python
sample_mean = []

for i in range(50):
    y = random.sample(x.tolist(), 4)
    avg = np.mean(y)
    sample_mean.append(avg)
```

```python
print(np.mean(sample_mean))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(sample_mean)
plt.show()
```
