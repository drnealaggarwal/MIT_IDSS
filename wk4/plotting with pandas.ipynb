```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
```

```python
fig, ax = plt.subplots()
ax.bar(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Dan's Nut Butter Store",
      ylabel="Price($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.tail(3)
```

```python
# histgram of age to see what the age data looks like
heart_disease["age"].plot.hist(bins=50);
```

```python
heart_disease.head(2)
```

```python
heart_disease.plot.hist(figsize=(10,10), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
# plotting directly using matplotlib's pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])
ax.set(title='Heart Disease and Cholesterol Level', xlabel='Age', ylabel='Cholesterol');
ax.legend(*scatter.legend_elements(), title='Target');
ax.axhline(over_50['chol'].mean(), linestyle='--');
```
